# #Thyroid prediction using Linear regression Modal

Used Linear regression modal to predict the Thyroid flag, Since linear regression modal is used we need to convert all the categorical values to numerical values in order to get proper prediction and also to avoid errors as numerical values are the ones this modal expects as input and the prediction output is also  numerical values.
Hence converted the prediction column Thyroid to numerical value as 0 and 1 and the same for the input values.
For sex, converted F,M,"?" to 0,1,2 and the remaining columns converted f and t to 0 and 1. Dropped the only row with ? of age column and Thyroid column P,N to 1 and 0

```python
import pandas as pd
from sklearn import linear_model
import pickle
import csv
import numpy as np
import sklearn
```

```python
df = pd.read_csv("thyroid.txt", sep="	")
```

```python
df
```

```python
df.age.unique() # to check the uniqueness of the column age
```

```python
df.drop(df[df.age=="?"].index, inplace=True) # to drop the row with the only ? value
```

```python
df['sex'].unique() # to check the uniqueness of the column sex
```

```python

df['sex'].replace({'F':1, 'M':0,'?':2}, inplace=True) # to replace the categorical value to numerical value to use it in the modal
    
```

```python
for i in ['on thyroxine','on antithyroid medication','sick','pregnant','thyroid surgery','I131 treatment','lithium','goitre','tumor','hypopituitary','psych']:
    df[i].replace({'f':1, 't':0}, inplace=True)  # to replace the categorical value to numerical value to use it in the train modal
```

```python
for i in ['TSH','T3','TT4','T4U','FTI']:
    df[i].replace({'?':0}, inplace=True) # to replace the categorical value to numerical value to use it in the train modal
```

```python
df["Thyroid"].unique() # to check the uniqueness of the column thyroid
```

```python
df["Thyroid"].replace({'P':1, 'N':0}, inplace=True) # to replace the categorical value to numerical value to use it in the test sample 
```

```python
#for i in ['TSH','T3','TT4','T4U','FTI','sex']:
for i in ['TSH','T3','TT4','T4U','FTI']:
    df.drop(df[df[i]=="?"].index, inplace=True) # to replace the categorical value to numerical value to use it in the train sample 
```

```python
# Defining Prediction and Test values
Predict = "Thyroid" # Assigned the column to be predicated to the predict variable for Y value assignment
```

```python
X = np.array((df.drop([Predict], 1))) #X and Y values assignment based on the columns to be used for prediction
Y = np.array((df[Predict]))
```

```python
#Pickle file created using the below code
# To store the best modal in the pickle file
best_scr = 0
for i in range(250):
    # Splitting TEST and TRAIN Samples
    X_Train, X_Test, Y_Train, Y_Test = sklearn.model_selection.train_test_split(X, Y, test_size=0.1)

    # Fetch Model
    Linear_mdl = linear_model.LinearRegression()

    # Fit Model
    Linear_mdl.fit(X_Train, Y_Train)

    # Check approx accuracy score of the fitted model
    acc_scr = Linear_mdl.score(X_Test, Y_Test)
    print(acc_scr)

    if acc_scr > best_scr:
        with open("geekschallange_JAI_VIGNESH_PR.pickle", "wb") as fle:
            pickle.dump(Linear_mdl, fle)
```
